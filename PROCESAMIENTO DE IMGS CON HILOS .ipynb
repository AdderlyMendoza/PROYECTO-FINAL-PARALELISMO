```python
# ----------------------------------------------- FUNCIONES PARA PROCESAR MÁGENES-------------------------------------------

def deteccion_borde(): # ALGORITMO 1
    global imagen_cargada
    imagen_cargada = imutils.resize(imagen_cargada, width=500)
    
    edge_img = cv2.Canny(imagen_cargada,100,200)
    
    im = Image.fromarray(edge_img)              # Image.fromarray = Crea una memoria de imagen a partir de un objeto
    img = ImageTk.PhotoImage(image=im)          # ImageTk.PhotoImage=  Una imagen fotográfica compatible con Tkinter
    
    label_1_img_procesada.configure(image=img)
    label_1_img_procesada.image = img
    
    
def rotacion(): # ALGORITMO 2
    global imagen_cargada
    
    imagen_cargada = imutils.resize(imagen_cargada, width=500)


    h, w = imagen_cargada.shape[:2]     # w = ANCHO, h = ALTO   = 500
    
    rot_matrix = cv2.getRotationMatrix2D((w/2,h/2), -180, 1) # -180 = ángulo en el que quieres rotar, 1 = escala
    rot_image = cv2.warpAffine(imagen_cargada, rot_matrix, (w, h))    # ROTANDO LA IMG
    
    imageToShow = imutils.resize(rot_image, width=500)
    imageToShow = cv2.cvtColor(imageToShow, cv2.COLOR_BGR2RGB)  # MOSTRAR LA IMG CON SUS MISMOS COLORES
    
    im = Image.fromarray(imageToShow)              # Image.fromarray = Crea una memoria de imagen a partir de un objeto
    img = ImageTk.PhotoImage(image=im)          # ImageTk.PhotoImage=  Una imagen fotográfica compatible con Tkinter
    
    label_2_img_procesada.configure(image=img)
    label_2_img_procesada.image = img


def eliminar_fondo(): # ALGORITMO 3
    global imagen_cargada
    imagen_cargada = imutils.resize(imagen_cargada, width=500)

    gray_img = cv2.cvtColor(imagen_cargada, cv2.COLOR_BGR2GRAY) #img escala grises

    _, thresh = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    img_contours = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[-2]
    
    img_contours = sorted(img_contours, key=cv2.contourArea)

    for i in img_contours:

        if cv2.contourArea(i) > 100:

            break
            
    mask = np.zeros(imagen_cargada.shape[:2], np.uint8)
    
    cv2.drawContours(mask, [i],-1, 255, -1)
    
    new_img = cv2.bitwise_and(imagen_cargada, imagen_cargada, mask=mask)
    
    new_img = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)  # MOSTRAR LA IMG CON SUS MISMOS COLORES
    
    im = Image.fromarray(new_img)              # Image.fromarray = Crea una memoria de imagen a partir de un objeto
    img2 = ImageTk.PhotoImage(image=im)          # ImageTk.PhotoImage=  Una imagen fotográfica compatible con Tkinter
    
    label_3_img_procesada.configure(image=img2)
    label_3_img_procesada.image = img2

    

def contraste(): # ALGORITMO 4
    
    def show_contraste(ev=None):
        global imagen_cargada

        contraste_img = imutils.resize(imagen_cargada, width=500) # tamaño
        contraste_img = Image.fromarray(contraste_img) # cambiar a tipo PIL

        contraste_img = transforms.ColorJitter(contrast=scale.get())(contraste_img)  # contraste_img = fuente

        img = ImageTk.PhotoImage(image=contraste_img)    # ImageTk.PhotoImage=  Una imagen fotográfica compatible con Tkinter

        label_4_img_procesada.configure(image=img)  # donde se mostrará
        label_4_img_procesada.image = img
        
        global save_img
        save_img = contraste_img

    scale = tk.Scale(panel_4_Info11, from_=0, to=100, orient=tk.HORIZONTAL,command=show_contraste)
    scale.set(1)
    scale.pack()

def escala_grises(): # ALGORITMO 5
    
    global imagen_cargada
    
    
    gray_img = imutils.resize(imagen_cargada, width=500)
    gray_img = cv2.cvtColor(gray_img, cv2.COLOR_BGR2GRAY)    #Usamos cvtColor para convertir a escala de grises
    
    im = Image.fromarray(gray_img)              # Image.fromarray = Crea una memoria de imagen a partir de un objeto
    img = ImageTk.PhotoImage(image=im)          # ImageTk.PhotoImage=  Una imagen fotográfica compatible con Tkinter
    
    label_5_img_procesada.configure(image=img)
    label_5_img_procesada.image = img
    

def saturacion():  # ALGORITMO 6
    
    def show_saturation(ev=None):
        global imagen_cargada

        contraste_img = imutils.resize(imagen_cargada, width=500) # tamaño
        contraste_img = Image.fromarray(contraste_img) # cambiar a tipo PIL

        contraste_img = transforms.ColorJitter(saturation=scale.get())(contraste_img) # contraste_img = fuente
                                                                                # saturation=scale.get() = para la saturación

        img = ImageTk.PhotoImage(image=contraste_img)    # ImageTk.PhotoImage=  Una imagen fotográfica compatible con Tkinter

        label_6_img_procesada.configure(image=img)  # donde se mostrará
        label_6_img_procesada.image = img
        
        global save_img
        save_img = contraste_img
        
    scale = tk.Scale(panel_6_Info11, from_=0, to=100, orient=tk.HORIZONTAL,command=show_saturation)
    scale.set(1)
    scale.pack()

def brillo():  # ALGORITMO 7
    
    def show_brillo(ev=None):
        global imagen_cargada

        contraste_img = imutils.resize(imagen_cargada, width=500) # tamaño
        contraste_img = Image.fromarray(contraste_img) # cambiar a tipo PIL

        contraste_img = transforms.ColorJitter(brightness=scale.get())(contraste_img) # contraste_img = fuente
                                                                                # saturation=scale.get() = para la saturación
            
        img = ImageTk.PhotoImage(image=contraste_img)    # ImageTk.PhotoImage=  Una imagen fotográfica compatible con Tkinter

        label_7_img_procesada.configure(image=img)  # donde se mostrará
        label_7_img_procesada.image = img
        
        global save_img
        save_img = contraste_img
        
    scale = tk.Scale(panel_7_Info11, from_=0, to=100, orient=tk.HORIZONTAL,command=show_brillo)
    scale.set(1)
    scale.pack()
    
    
def extraer_texto():  # ALGORITMO 8
    
    # RUTA DONDE ESTÁ NUESTRA TESSERACT 
    # SI NO LE EJECUTA ESTA LÍNEA ver vídeo: https://www.youtube.com/watch?v=_j6o2rBkhhg
    pytesseract.pytesseract.tesseract_cmd = r'C:\Users\avantec\AppData\Local\Programs\Tesseract-OCR\Tesseract'

    global imagen_cargada   

    texto = pytesseract.image_to_string(imagen_cargada)  # METODO "image_to_string" para convertir la img

    #print("asdasdasdasda: ",texto)
    
    label_texto = tk.Label(panel_8_Info9, text=texto,font=("Arial",16,"bold"))
    label_texto.pack()

    
def reducir_ruido():   # ALGORITMO 9
    
    global imagen_cargada
    
    img_resultado = imutils.resize(imagen_cargada, width=500)
    img_resultado = cv2.fastNlMeansDenoisingColored(img_resultado,None,20,10,7,21)   # RUIDO
    
    im = Image.fromarray(img_resultado)              # Image.fromarray = Crea una memoria de imagen a partir de un objeto
    img = ImageTk.PhotoImage(image=im)          # ImageTk.PhotoImage=  Una imagen fotográfica compatible con Tkinter
    
    label_9_img_procesada.configure(image=img)
    label_9_img_procesada.image = img
    
    
def deteccion_color():

    def deteccion_color_1():
        global imagen_cargada
        imagen_cargada = imutils.resize(imagen_cargada, width=500)
        image = imagen_cargada
        if selected.get() == 1:
            # Rojo
            rangoBajo1 = np.array([0, 140, 90], np.uint8)
            rangoAlto1 = np.array([8, 255, 255], np.uint8)
            rangoBajo2 = np.array([160, 140, 90], np.uint8)
            rangoAlto2 = np.array([180, 255, 255], np.uint8)
        if selected.get() == 2:
            # Amarillo
            rangoBajo = np.array([10, 98, 0], np.uint8)
            rangoAlto = np.array([25, 255, 255], np.uint8)

        imageGray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        imageGray = cv2.cvtColor(imageGray, cv2.COLOR_GRAY2BGR)
        imageHSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        if selected.get() == 1:
            # Detectamos el color rojo
            maskRojo1 = cv2.inRange(imageHSV, rangoBajo1, rangoAlto1)
            maskRojo2 = cv2.inRange(imageHSV, rangoBajo2, rangoAlto2)
            mask = cv2.add(maskRojo1, maskRojo2)
        else:
            # Detección para el color Amarillo y Azul celeste
            mask = cv2.inRange(imageHSV, rangoBajo, rangoAlto)
        mask = cv2.medianBlur(mask, 7)
        colorDetected = cv2.bitwise_and(image, image, mask=mask)
        # Fondo en grises
        invMask = cv2.bitwise_not(mask)
        bgGray = cv2.bitwise_and(imageGray, imageGray, mask=invMask)
        # Sumamos bgGray y colorDetected
        finalImage = cv2.add(bgGray, colorDetected)
        imageToShowOutput = cv2.cvtColor(finalImage, cv2.COLOR_BGR2RGB)
        

        
        # Para visualizar la imagen en lblOutputImage en la GUI
        im = Image.fromarray(imageToShowOutput)
        img = ImageTk.PhotoImage(image=im)
        label_10_img_procesada.configure(image=img)
        label_10_img_procesada.image = img
        # Label IMAGEN DE SALIDA
        #lblInfo3 = Label(root, text="IMAGEN DE SALIDA:", font="bold")
        #lblInfo3.grid(column=1, row=0, padx=5, pady=5)
    
    selected = IntVar()
    rad1 = Radiobutton(panel_10_Info11, text='Rojo', width=20,value=1, variable=selected, command= deteccion_color_1)
    rad2 = Radiobutton(panel_10_Info11, text='Amarillo',width=20, value=2, variable=selected, command= deteccion_color_1)
    rad1.grid(column=0, row=0)
    rad2.grid(column=1, row=0)
```
